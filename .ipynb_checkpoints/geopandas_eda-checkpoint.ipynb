```python
# Import necessary modules
import geopandas as gpd
import matplotlib as plt
from matplotlib.pyplot import figure
import geoplot
import numpy as np
import pandas as pd
```

```python

# Set filepath (fix path relative to yours)
fp = "./data/Collisions/Collisions.shp"

# Read file using gpd.read_file()
cl = gpd.read_file(fp)

#Read file using gpd.read_file()
fp2 = "./data/Seattle_Streets/Seattle_Streets.shp"
street = gpd.read_file(fp2)

fp3 = "./data/Sidewalks/Sidewalks.shp"
sidewalk = gpd.read_file(fp3)


fp4 = "./data/Existing_Bike_Facilities/Existing_Bike_Facilities.shp"
bikelane = gpd.read_file(fp4)
```

```python
# filter out highway, 0 is non-artisial 
strt = street.loc[street['ARTCLASS'] < 4,].reset_index(drop = True).copy()

## remove road underconstruction or planned
strt2 = strt.loc[(strt['STATUS'] != 'UNDERCONS') & (strt['STATUS'] != 'PLANNED'),:].reset_index(drop = True).copy()
```

```python
# filter our bike collision
bikecl_ = cl.loc[cl['PEDCYLCOUN'] >=1,]
bikecl = bikecl_.reset_index(drop = True).copy()
len(bikecl)
```

```python
min_length = np.empty (len(bikecl))
min_index = np.empty (len(bikecl))
for i in np.arange(0,len(bikecl)): 
    try:
        distance = [bikecl.iloc[i,-1].distance(strt2.iloc[j,-1]) for  j in np.arange(0,len(strt2))]
        min_length[i] = min(distance)
        min_index[i] = distance.index(min(distance))
        print(i)
    except: 
        min_length[i] = np.NaN
        min_index[i] = np.NaN
        
```

```python
len(min_index)
len(min_length)
len(bikecl)
```

```python
bikecl['near_str_index'] = min_index
bikecl['min_length'] = min_length
```

```python
weather_group = np.empty(len(bikecl), dtype = object)
for i in np.arange(0,len(bikecl['WEATHER'])):
        weather = bikecl['WEATHER'][i] 
        if ( weather == 'Blowing Sand or Dirt or Snow') or (weather == 'Snowing') or (weather == 'Sleet/Hail/Freezing Rain') or (weather == 'Snowing') or (weather == 'Fog/Smog/Smoke') :
            weather_group[i] = 'Raining'
        elif (weather == 'Unknown') or (weather == 'Other'):
            weather_group[i] = np.nan
        else:
            weather_group[i] = weather

    
```

```python
bikecl['weather_group'] = weather_group
```
